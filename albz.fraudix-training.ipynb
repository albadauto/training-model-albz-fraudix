```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('./db/db.csv')
```

```python
dataset
```

```python
dataset['Valor_Reembolso'] = dataset['Valor_Reembolso'].str.replace(',', '.')
dataset['Distancia_Percorrida'] = dataset['Distancia_Percorrida'].str.replace(',', '.')
```

```python
dataset
```

```python
X_train = dataset[['Valor_Reembolso', 'Distancia_Percorrida', 'Num_Reembolsos_Recentes']]
```

```python
y_train = dataset[['Probabilidade_Fraude']]
```

```python
y_train.shape
```

```python
X_train.shape
```

```python
dataset[dataset['Probabilidade_Fraude'] != '1']
```

```python
y_train['Probabilidade_Fraude'] = y_train['Probabilidade_Fraude'].str.replace(',', '.').astype(float)

# Converter todas as colunas de X_train para float
X_train = X_train.astype(float)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=[3]),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=16, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid'),
])

model.compile(optimizer=tf.keras.optimizers.Adam(), loss='binary_crossentropy')
```

```python
y_train[y_train['Probabilidade_Fraude'] != '0']
```

```python
model.fit(X_train, y_train, epochs=50)
```

```python
X_train
```

```python
import numpy as np
predict = model.predict(np.array([[200, 200, 1]]))
```
